```python
import pandas as pd
import glob
import os

# Tentukan path folder
folder1 = 'SteamReviews'
folder2 = 'SteamReviews_2'

# Ambil semua file csv dari kedua folder
csv_files_folder1 = glob.glob(os.path.join(folder1, '*.csv'))
csv_files_folder2 = glob.glob(os.path.join(folder2, '*.csv'))

# Gabungkan semua path csv ke dalam satu list
all_csv_files = csv_files_folder1 + csv_files_folder2

# Buat list untuk menampung DataFrame sementara
dataframes = []

# Baca dan gabungkan semua csv
for file in all_csv_files:
    df = pd.read_csv(file)
    dataframes.append(df)

# Gabungkan semua DataFrame jadi satu
combined_df = pd.concat(dataframes, ignore_index=True)

# Simpan ke file CSV baru (opsional)
combined_df.to_csv('combined_reviews.csv', index=False)
```

```python
# Membaca file CSV
df = pd.read_csv('combined_reviews.csv')

# Mengambil 50 sampel secara acak
sample_df = df.sample(n=50, random_state=42)  # random_state untuk hasil yang bisa direproduksi

# Menampilkan hasil sampel
print(sample_df)

sample_df.to_csv('sampled_reviews_50.csv')
```

```python
import pandas as pd

# Baca file
df = pd.read_csv('combined_reviews.csv')

# Hitung jumlah review per UserID
user_counts = df['UserID'].value_counts()

# Ambil UserID yang punya minimal 5 review
eligible_users = user_counts[user_counts >= 5].index

# Ambil 20 user secara acak dari yang memenuhi syarat
selected_users = pd.Series(eligible_users).sample(n=20, random_state=42)

# Ambil 5 review acak dari masing-masing user
sampled_reviews = (
    df[df['UserID'].isin(selected_users)]
    .groupby('UserID')
    .apply(lambda x: x.sample(n=5, random_state=42))
    .reset_index(drop=True)
)

# Cek hasil
print(sampled_reviews)

# (Opsional) Simpan ke file
sampled_reviews.to_csv('100_sampled_reviews.csv', index=False)
```
